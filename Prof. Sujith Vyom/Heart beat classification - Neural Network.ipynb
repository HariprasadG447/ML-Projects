```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
import tensorflow as tf
```

```python
train_data = pd.read_csv('/content/drive/MyDrive/Kaggle ECG data/mitbih_train.csv',header=None)
test_data = pd.read_csv('/content/drive/MyDrive/Kaggle ECG data/mitbih_test.csv',header=None)
```

```python
train_data
```

#Shuffling the data for better results

```python
training_data = train_data.sample(frac = 1).reset_index(drop = True)
training_data
```

```python
test_set = test_data.sample(frac = 1).reset_index(drop = True)
```

### As the last row are the categories, we take the last column as the y parameter of the model

```python
x_train = training_data.drop(187,axis = 1)
y_train = training_data[187].copy()
```

```python
y_train
```

```python
x_train
```

```python
x_test = test_set.drop(187,axis = 1)
y_test = test_set[187].copy()
```

#Training

```python
inputs = tf.keras.Input(shape = (x_train.shape[1]))
x = tf.keras.layers.Dense(64, activation = 'relu')(inputs)
x = tf.keras.layers.Dense(64, activation = 'relu')(x)

outputs = tf.keras.layers.Dense(5,activation = 'softmax')(x)

model = tf.keras.Model(inputs = inputs, outputs = outputs)
```

```python
model.compile(optimizer = 'adam', loss = tf.keras.losses.SparseCategoricalCrossentropy(), metrics = ['accuracy', tf.keras.metrics.SparseCategoricalAccuracy()])

history = model.fit(
    x_train,
    y_train,
    validation_split = 0.2,
    batch_size = 32, epochs = 100,
    callbacks = [tf.keras.callbacks.EarlyStopping(monitor = 'val_loss', patience = 3, restore_best_weights = True)]
)
```

# Results

```python
results = model.evaluate(x_test,y_test)
```

```python
print('Test set accuracy : {:.4f}%'.format(results[1] * 100))
print('Test set sparse categorical accuracy : {:.4f}%'.format(results[2] * 100))
```
